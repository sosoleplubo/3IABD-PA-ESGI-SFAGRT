## Classification

### Linear Simple :

        MLP (2, 1)   : OK

```python
import PAESGISFAGRT

X = np.array([
      [1, 1],
      [2, 3],
      [3, 3]
])
Y = np.array([
      1,
      -1,
      -1
])

dataset = list(zip(X.tolist(), [[y] for y in Y.tolist()]))

model = PAESGISFAGRT.ModelePMC(
    couches=[2, 1],
    txapprentissage=0.01,
    activation="tanh",
    probleme="classification",
)

model.entrainer(dataset, nbgeneration=100000)

for x in X.tolist():
    print(model.predire(x))
```

        Linear Model    : KO
        MLP (2, 4, 1)   : OK

```python
import PAESGISFAGRT

X = np.random.random((500, 2)) * 2.0 - 1.0
Y = np.array([1 if abs(p[0]) <= 0.3 or abs(p[1]) <= 0.3 else -1 for p in X])

plt.scatter(X[Y == 1, 0], X[Y == 1, 1], color='blue')
plt.scatter(X[Y == -1, 0], X[Y == -1, 1], color='red')
plt.show()
plt.clf()

dataset = list(zip(X.tolist(), [[y] for y in Y.tolist()]))

test_points = [
    [0.0, 0.0],   #1
    [0.1, 0.5],   #1
    [0.8, 0.8],   #-1
    [-0.2, 0.9],  #-1
    [-0.7, -0.7], #-1
]

model = PAESGISFAGRT.ModelePMC(
    couches=[2,4,1],
    txapprentissage=0.01,
    activation="tanh",
    probleme="classification",
)

model.entrainer(dataset, nbgeneration=100000)

for p in test_points:
    pred = model.predire(p)
    print(pred)
```

```python
plt.scatter(np.array(list(map(lambda elt : elt[1], filter(lambda c: Y[c[0]] == 1, enumerate(X)))))[:,0], np.array(list(map(lambda elt : elt[1], filter(lambda c: Y[c[0]] == 1, enumerate(X)))))[:,1], color='blue')
plt.scatter(np.array(list(map(lambda elt : elt[1], filter(lambda c: Y[c[0]] == -1, enumerate(X)))))[:,0], np.array(list(map(lambda elt : elt[1], filter(lambda c: Y[c[0]] == -1, enumerate(X)))))[:,1], color='red')
plt.show()
plt.clf()
```

### Multi Linear 3 classes :
        Linear Model x3 : OK
        MLP (2, 3)      : OK

```python
X = np.random.random((500, 2)) * 2.0 - 1.0
Y = np.array([[1, -1, -1] if -p[0] - p[1] - 0.5 > 0 and p[1] < 0 and p[0] - p[1] - 0.5 < 0 else
              [-1, 1, -1] if -p[0] - p[1] - 0.5 < 0 and p[1] > 0 and p[0] - p[1] - 0.5 < 0 else
              [-1, -1, 1] if -p[0] - p[1] - 0.5 < 0 and p[1] < 0 and p[0] - p[1] - 0.5 > 0 else
              [-1, -1, -1]for p in X])

X = X[[not np.all(arr == [-1, -1, -1]) for arr in Y]]
Y = Y[[not np.all(arr == [-1, -1, -1]) for arr in Y]]

dataset = list(zip(X.tolist(), Y.tolist()))

model = PAESGISFAGRT.ModelePMC(
    couches=[2, 3],
    txapprentissage=0.01,
    activation="tanh",
    probleme="classification",
)

model.entrainer(dataset, nbgeneration=10000)

test_points = [
    [-0.5, -0.3],
    [0.2, 0.4],
    [0.6, -0.3],
]

for p in test_points:
    pred = model.predire(p)
    print(f"{p} → {pred}")
```

```python
plt.scatter(np.array(list(map(lambda elt : elt[1], filter(lambda c: Y[c[0]][0] == 1, enumerate(X)))))[:,0], np.array(list(map(lambda elt : elt[1], filter(lambda c: Y[c[0]][0] == 1, enumerate(X)))))[:,1], color='blue')
plt.scatter(np.array(list(map(lambda elt : elt[1], filter(lambda c: Y[c[0]][1] == 1, enumerate(X)))))[:,0], np.array(list(map(lambda elt : elt[1], filter(lambda c: Y[c[0]][1] == 1, enumerate(X)))))[:,1], color='red')
plt.scatter(np.array(list(map(lambda elt : elt[1], filter(lambda c: Y[c[0]][2] == 1, enumerate(X)))))[:,0], np.array(list(map(lambda elt : elt[1], filter(lambda c: Y[c[0]][2] == 1, enumerate(X)))))[:,1], color='green')
plt.show()
plt.clf()
```

### Multi Cross :
